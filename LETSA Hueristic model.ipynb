```python
import pandas as pd
import numpy as np

# Read the data
df = pd.read_excel("testing.xlsx")

# Initialize parameters
n = len(df["Node Index"])
nf = 1  # Number of final assemblies
final_job = n  # job number of final assembly, should correspond to nf
m = 2  # Number of machines
print(n, final_job, m)

# Initialize lists and variables
feasible_job_ls = [int(job) for job in df["Node Index"]]  # List of feasible jobs F
ops_schedule = []  # Operations/job schedule
end_times = []  # End time for each operation, corresponds to ops schedule
machine_end_times = [0] * m  # Track end time for each machine
makespan = 0

# Helper function to get predecessors
def get_predecessors(index):
    preds = df["Predecessors"][index]
    if pd.isna(preds):
        return []
    else:
        return list(map(int, preds.split(",")))

while feasible_job_ls:
    current_job = feasible_job_ls.pop(0)
    index = int(current_job) - 1
    predecessors = get_predecessors(index)
    processing_time = int(df["Processing Time"][index])
    
    # Check if all predecessors are in the ops_schedule
    if all(pred in ops_schedule for pred in predecessors):
        # Find a machine that is free (earliest available time)
        machine_index = machine_end_times.index(min(machine_end_times))
        
        # Calculate the start time for this job
        start_time = machine_end_times[machine_index]
        end_time = start_time + processing_time
        machine_end_times[machine_index] = end_time
        
        # Update the makespan
        makespan = max(makespan, end_time)
        
        # Schedule the job
        ops_schedule.append(current_job)
        end_times.append(end_time)
    else:
        # Re-add the job to the end of the list if it can't be scheduled yet
        feasible_job_ls.append(current_job)

print(ops_schedule)
print(end_times)
print(makespan)
```
